##### 스택으로 수열 만들기

- 계산 방법



```python
import time

n = int(input())
# A = list(map(int, input().split()))
ZA = [0] * n

for i in range(n):
    A[i] = int(input())
    
# print(n, A)

stack = []    # 스택으로 사용할 리스트
num = 1       # 1부터 n까지 오름차순으로 push할 변수(자연수)
result = True
answer=''

sTime = time.time()
for i in range(n):
    suNo = A[i]    # 수열의 i번째 수
    if suNo >= num:
        while suNo >= num:
            stack.append(num)  # 스택에 push
            num += 1           # 다음 수를 push하기 위해 1 증가(자연수 증가)
            answer += '+\n'    # push했으므로 + 출력
        stack.pop()
        answer += '-\n'
    else:
        n = stack.pop()    # 스택의 최상위 원소를 꺼냄
        if n > suNo:
            print('NO')
            result = False
            break
        else:
            answer += '-\n'
            
print(time.time() - sTime)       

if result:
    print(answer)

```

#### 큐를 이용한 카드게임(p86)

```python
# 코드 구현
# deque: 양방향 큐 구현을 위한 라이브러리
# 일반적인 리스트(list)가 이러한 연산에 O(n)이 소요되는 데 반해, 데크(deque)는 O(1)로 접근 가능
from collections import deque 
import time

n = int(input())
myQueue = deque()

for i in range(1, n+1):
    myQueue.append(i)

sTime = time.time()

while len(myQueue) > 1:  #  카드가 1개 남을 때까지 반복
    myQueue.popleft()    #  맨 앞의 카드를 버림
    myQueue.append(myQueue.popleft())  #  맨 앞의 카드를 맨 뒤로 옮김

print(time.time() - sTime)  
print(myQueue[0])
```

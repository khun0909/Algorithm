```python
#  
# deque:     
#  (list)   O(n)   , (deque) O(1)  
from collections import deque 
import time

n = int(input())
myQueue = deque()

for i in range(1, n+1):
    myQueue.append(i)

sTime = time.time()

while len(myQueue) > 1:  #   1   
    myQueue.popleft()    #     
    myQueue.append(myQueue.popleft())  #       

print(time.time() - sTime)  
print(myQueue[0])
```

```python
# 코드 구현, O(N)의 시간 복잡도를 가지는 투 포인터 알고리즘 예제
import time

n = int(input())   # 데이터의 개수 N
cnt = 1
sindex = 1
eindex = 1
Sum = 1

start_time=time.time()
while eindex != n:
    if Sum == n:
        cnt += 1
        eindex += 1
        Sum += eindex
    elif Sum > n:
        Sum -= sindex
        sindex += 1
    else:
        eindex += 1
        Sum += eindex
        
end_time=time.time()
     
print(cnt, end_time-start_time)
```

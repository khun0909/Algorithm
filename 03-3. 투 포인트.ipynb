```python
#  , O(N)       
import time

n = int(input())   #   N
cnt = 1
sindex = 1
eindex = 1
Sum = 1

start_time=time.time()
while eindex != n:
    if Sum == n:
        cnt += 1
        eindex += 1
        Sum += eindex
    elif Sum > n:
        Sum -= sindex
        sindex += 1
    else:
        eindex += 1
        Sum += eindex
        
end_time=time.time()
     
print(cnt, end_time-start_time)
```

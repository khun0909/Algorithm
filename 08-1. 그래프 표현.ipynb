```python
#  

from collections import deque #  

import sys
# input = sys.stdin.readline

n, m, k, x = map(int, input().split()) #  ,  ,  ,   
data = [[] for _ in range(n+1)] #   1  n+1
answer = [] #   k   
visited = [-1] * (n+1) #    

# DFS  (  )
def DFS(x):
    queue = deque() #   
    queue.append(x) #     
    visited[x] += 1 #    
    
    while queue: #    
        new_node = queue.popleft() #    
        for i in data[new_node]: #      
            if visited[i] == -1:
                queue.append(i) #     
                visited[i] = visited[new_node] + 1 #  


#     
for _ in range(m):
    a, b = map(int, input().split())
    data[a].append(b) # a  b    
    
DFS(x) # DFS  

#   k  answer  
for i in range(1, n+1):
    if visited[i] == k:
        answer.append(i)
        
#  
if not answer:  #   k   -1 
    print(-1)
else:           #   k     
    answer.sort()
    for i in answer:
        print(i)
```

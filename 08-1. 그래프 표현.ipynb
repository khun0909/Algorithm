##### 그래프 구조
- vertex (정점) (node)과 edge (간선) 으로 이루어진 <b>자료구조</b>다. (비선형 구조)
- degree (차수): node에 연결된 edge의 수. 예: A의 차수 1 B의 차수 2



- 경로
    -Node 표현: A - B - C
    - Edge 표현: (A, B) (B, C)
        - 단순 경로 : node를 최대 한 번만 지나는 경로
        - Cycle : 시작과 끝의 node가 같은 경로
        - DAG(Directed Acycle Graph)

- 두 개의 정점(node)이 연결된 상태를 인접하다고 한다.<br> 
(Directed graph에서는 방향이 향하는 쪽 node가 인접한 정점이다.) 
- 완전 그래프: 모두 Adjacency한 그래프



##### 엣지 리스트(edge list)
- 엣지를 중심으로 그래프를 표현한다.
- 엣지 리스트는 리스트에 출발 노드, 도착 노드를 저장하여 옛지를 표현한다. 
- 또는 출발 노드, 도착 노드, 가중치를 저장하여 가중치가 있는 엣지를 표현한다. 

##### 엣지 리스트로 가중치 없는 그래프 표현하기 
- 가중치가 없는 그래프는 출발 노드와 도착 노드만 표현하므로 리스트의 열은 2개면 충분하다.



- 1에서 2로 뻗어나가는 에지는 [1, 2]로 표현
- 4에서 5로 뻗어나가는 에지는 [4, 5]로 표현
- 이처럼 방향이 있는 그래프는 순서에 맞게 노드를 리스트에 저장하는 방식으로 표현.<br>
그리고 노드를 리스트에 저장하여 엣지를 표현하므로 엣지 리스트라 한다.

##### 엣지 리스트로 가중치 있는 그래프 표현하기
- 가중치가 있는 그래프는 열을 3개로 늘려 3번째 열에 가중치를 저장하면 된다.



- 1에서 2로 향하는 가중치가 8인 엣지는 이제 [1, 2, 8]로 표현
- 이처럼 엣지 리스트는 구현하기가 쉽다.

- 하지만 특정 노드와 관련되어 있는 엣지를 탐색하기는 쉽지 않다.<br> 
에지 리스트는 노드 사이의 최단 거리를 구하는 벨만-포드나 최소 신장 트리를 찾는 크루스칼 알고리즘에 사용하며,<br> 
노드 중심 알고리즘에는 잘 사용하지 않는다.

##### 인접 행렬(adjacency matrix)
- 2차원 리스트를 자료구조로 이용하여 그래프를 표현한다.
- Vertex * Vertex 크기 만큼 행렬을 만들어서, 그 두개의 Vertex가 Edge로 이어져있다면 1 , 아니면 0을 넣어 그래프를 표현하는 방식이다.
- 인접 행렬로 구현을 할 경우, 그래프(무방향의 한해서) 대각 성분을 기준으로 대칭인 성질을 갖게된다

- 인접 행렬로 가중치 없는 그래프 표현하기
    - 다음 그림을 보면 1에서 2를 향하는 에지를 인접 행렬은 1행 2열에 1을 저장하는 방식으로 표현
    - 1을 저장하는 이유는 가중치가 없기 때문이다. 
    - 1에서 2로 향하는 엣지가 있다는 표시를 노드 중심으로 한다고 이해하면 된다.



- 인접 행렬로 가중치 있는 그래프 표현하기
    - 2에서 5로 향하는 에지의 가중치를 2행 5열에 기록.



- 장점
    - 구현이 간편하다. 
    - Vertex끼리 연결되어 있는지 확인하고 싶으면, adj[i][j]만 확인하면 되기때문에, 탐색(접근)이 빠르다.
- 단점
    - 메모리가 많이 듭니다. 극단적으로, 정점(vertex)가 1,000,000개면 , 1,000,000*1,000,000개의 배열을 만들어야 한다.
    - 노드에 비해 간선이 너무 적으면 비효율적이게 됨. 예를들어, 극단적으로 노드가 1,000,000개인데, 내가 간선 1개만 갖고 있으면, <br>
    최악의 경우, 1,000,000번 탐색을 해야할 수 도 있다는 이야기다.

##### 인접 리스트(adjacency list)
- 파이썬의 리스트를 이용하여 그래프를 표현. 
- 노드 개수만큼 리스트를 선언합니다. 리스트의 input data 형태는 문제의 조건에 맞게 설정한다.<br><br>
- 인접 리스트로 가중치 없는 그래프 표현하기
    - 현재의 경우 int 데이터 하나로 그래프를 표현하기에 충분하다. 
    - 인접 리스트에는 N번 노드와 연결되어 있는 노드를 리스트의 index N에 연결된 노드 개수만큼 리스트에 append하는 <br>
    방식으로 표현합니다.



- 예를 들어 노드 1과 연결된 2, 3 노드는 A[1]에 [2, 3]을 연결하는 방식으로 표현한다.

- 인접 리스트로 가중치 있는 그래프 표현하기
- 가중치가 있는 경우 input data를 2개(도착 노드, 가중치)로 사용한다. 
- 다음은 (도착 노드, 가중치)를 갖는 input data의 형태로 인접 리스트를 표현한 것이다.



- A[1]에 [(2, 8), (3, 3)]이 연결되어 있다. 
- 이는 노드 1과 2가 가중치 8 엣지로, 노드 1과 3이 가중치 3 엣지로 연결되어 있다는 것을 보여 준다. 
- 방향성도 고려 되었다. 그래프를 구현하는 다른 방법에 비해 인접 리스트를 이용한 그래프 구현은 복잡한 편입니다. 
- 하지만 노드와 연결된 엣지를 탐색하는 시간은 매우 뛰어나며, 노드 개수가 커도 공간 효율이 좋아 메모리 초과 에러도 발생하지<br>
않는다. 이런 장점으로 실제 코딩 테스트에서는 인접 리스트를 이용한 그래프 구현을 선호한다.

#### [문제] 특정 거리의 도시 찾기



##### 문제 분석하기
- 모든 도로의 거리가 1이므로 가중치가 없는 인접 리스트로 이 그래프를 표현할 수 있다. 
- 도시의 개수가 300,000, 도로의 최대 크기가 1,000,000이므로 BFS 탐색을 수행하면 이 문제를 시간 복잡도 안에서 해결할 수 있습니다.

##### [문제 풀이]
1. 인접 리스트로 도시와 도로 데이터의 그래프를 구현합니다.



2. BFS 탐색 알고리즘으로 탐색을 수행하면서 각 도시로 가는 최단 거릿값을 방문 리스트에 저장



    - 최초에는 방문 도시가 1이고, 이동하지 않았으므로 방문 리스트에 0을 저장합니다. 
    - 이후 방문하는 도시는 이전 도시의 방문 리스트값 + 1을 방문 리스트에 저장하는 방식으로 이동 거리를 저장합니다. 

3. 탐색 종료 후 방문 리스트에서 값이 K와 같은 도시의 번호를 출력



```python
# 코드 구현

from collections import deque # 양방향 큐

import sys
# input = sys.stdin.readline

n, m, k, x = map(int, input().split()) # 도시 개수, 도로 개수, 거리 정보, 출발 도시 번호
data = [[] for _ in range(n+1)] # 도시 번호는 1번부터 시작하므로 n+1
answer = [] # 최단 거리가 k인 도시들을 저장할 리스트
visited = [-1] * (n+1) # 방문 여부를 저장할 리스트

# DFS 함수 정의(방문 리스트 작성)
def DFS(x):
    queue = deque() # 양방향 큐 생성
    queue.append(x) # 시작 노드를 큐에 삽입 
    visited[x] += 1 # 시작 노드 방문 처리
    
    while queue: # 큐가 빌 때까지 반복
        new_node = queue.popleft() # 큐에서 노드를 하나 꺼내옴
        for i in data[new_node]: # 꺼내온 노드와 연결된 노드들을 확인 
            if visited[i] == -1:
                queue.append(i) # 방문하지 않은 노드라면 큐에 삽입
                visited[i] = visited[new_node] + 1 # 방문 처리


# 도로개수만큼 반복하여 도로 정보를 입력받음
for _ in range(m):
    a, b = map(int, input().split())
    data[a].append(b) # a번 노드와 b번 노드가 연결되어 있음을 의미
    
DFS(x) # DFS 함수 호출

# 최단 거리가 k인 도시들을 answer 리스트에 삽입
for i in range(1, n+1):
    if visited[i] == k:
        answer.append(i)
        
# 답안 출력
if not answer:  # 최단 거리가 k인 도시가 없다면 -1 출력
    print(-1)
else:           # 최단 거리가 k인 도시가 있다면 오름차순으로 정렬후 출력
    answer.sort()
    for i in answer:
        print(i)
```

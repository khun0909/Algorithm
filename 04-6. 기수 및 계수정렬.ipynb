```python
#  
from collections import deque

def radix_sort(nums):
    tmps = [deque() for _ in range(10)]  # 0 ~ 9    deque 
    
    maxVal = max(nums)  #  Z
    Q = deque(nums)  #  deque 
    cur_ten = 1  #   
    
    while maxVal >= cur_ten:
        while Q:
            num = Q.popleft()  #    
            tmps[ (num//cur_ten) % 10 ].append(num)  #    tmp 
            
        for tmp in tmps:
            while tmp:
                Q.append(tmp.popleft())  # tmp    
        
        cur_ten *= 10  #  
        
    return list(Q)

nums =[16, 80, 18, 77, 3, 24, 88, 23]
print(radix_sort(nums))
```

```python
#  

def counting_sort(nums):
    tmps = [0] * (max(nums) + 1)  # nums  + 1   
    
    for num in nums:
        tmps[num] += 1
        
    for i in range(len(tmps)):
        if tmps[i] != 0:
            for j in range(tmps[i]):
                print(i, end=' ')
                
    print(tmps)
 
   
nums=list(map(int,input().split()))
counting_sort(nums)
```

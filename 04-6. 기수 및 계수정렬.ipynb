##### 기수 정렬(radix sort)
- 값을 비교하지 않는 특이한 정렬
- 기수 정렬은 값을 놓고 비교할 자릿수를 정한 다음 해당 자릿수만 비교한다. 
- 기수 정렬의 시간 복잡도는 O(kn)으로, 여기서 k는 데이터의 자릿수를 말한다.



##### 수 정렬하기 3 (계수 정렬)



```python
# 코드 구현
from collections import deque

def radix_sort(nums):
    tmps = [deque() for _ in range(10)]  # 0 ~ 9 까지의 숫자를 담을 deque 생성
    
    maxVal = max(nums)  # 최대값을 구함Z
    Q = deque(nums)  # 입력값을 deque에 담음
    cur_ten = 1  # 자릿수를 나타내는 변수
    
    while maxVal >= cur_ten:
        while Q:
            num = Q.popleft()  # 큐에서 값을 하나 꺼냄
            tmps[ (num//cur_ten) % 10 ].append(num)  # 해당 자릿수의 숫자를 tmp에 담음
            
        for tmp in tmps:
            while tmp:
                Q.append(tmp.popleft())  # tmp에 담긴 값을 큐에 담음
        
        cur_ten *= 10  # 자릿수를 올림
        
    return list(Q)

nums =[16, 80, 18, 77, 3, 24, 88, 23]
print(radix_sort(nums))
```

#### 계수(카운팅) 정렬
- 크기를 기준으로 갯수를 세는 것을 통해 정렬하는 알고리즘
- 비교를 하는 것이 아니기 때문에 위치를 바꿔주는 과정 등이 필요없어 코드가 간결하다.
- O(n+k) 

- 풀이 방법



```python
# 코드 구현

def counting_sort(nums):
    tmps = [0] * (max(nums) + 1)  # nums의 최댓값 + 1 길이의 리스트 생성
    
    for num in nums:
        tmps[num] += 1
        
    for i in range(len(tmps)):
        if tmps[i] != 0:
            for j in range(tmps[i]):
                print(i, end=' ')
                
    print(tmps)
 
   
nums=list(map(int,input().split()))
counting_sort(nums)
```

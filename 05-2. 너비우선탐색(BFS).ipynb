```python
#  
from collections import deque

# N, M     
N, M = map(int, input().split())

# 2     
graph = []
for i in range(N):
    graph.append(list(map(int, input())))

#      (, , , )
dx = [-1, 1, 0, 0] # 
dy = [0, 0, -1, 1] # 

# BFS  
def bfs_q(x, y):
    # (Queue)   deque  
    queue = deque()
    queue.append((x, y)) #       
    
    #    
    while queue:
        x, y = queue.popleft()
        #   4   
        for i in range(4): # 
            nx = x + dx[i]
            ny = y + dy[i]
        
            #      
            if nx < 0 or ny < 0 or nx >= N or ny >= M:
                continue
                        
            # ()  
            if graph[nx][ny] == 0:
                continue
            
            #        
            if graph[nx][ny] == 1: #   
                graph[nx][ny] = graph[x][y] + 1 #    + 1
                queue.append((nx, ny)) #            
            
    #      
    return graph[N - 1][M - 1]
    

# BFS   
print(bfs_q(0, 0))
```

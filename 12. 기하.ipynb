##### 기하 알고리즘
- 기하 알고리즘은 점, 선, 다각형, 원과 같이각종 기하학적 도형을 다루는 알고리즘
- 하지만 실제 코딩 테스트에서의 기하 알고리즘은 모두 선분교차판별(CCW)라는 기하학적 성질을 이용해 문제를 풀이한다.

##### 선분교차판별(CCW:counter-clockwise)
- 평면상의 3개의 점과 관련된 점들의 위치 관계를 판단하는 알고리즘
- 세 점의 방향성



- 다음 그림을 보면 CCW값이 음수여도 교차하지 않는 경우이다.





- 다음과 같은 경우에는 A,B와 C,D의 상대적 위치를 비교해 줘야한다. 
- A,B둘중에 좌표가 큰값과 C,D중에 좌표가 큰값을 서로의 작은 값보다 클 때만 선분이 교차한다.

- 기타: |CCW 결괏값| / 2는 세 점으로 이뤄진 삼각형의 넓이로 이해하면 된다.

##### [문제] 선분 방향 구하기



##### 풀이



```python
# 코드 입력
x1, y1 = map(int, input().split())   # 첫 번째 점의 좌표
x2, y2 = map(int, input().split())   # 두 번째 점의 좌표
x3, y3 = map(int, input().split())   # 세 번째 점의 좌표

# CCW 공식에 따라 계산
result = (x1*y2 + x2*y3 + x3*y1) - (x2*y1 + x3*y2 + x1*y3)

if result > 0:
    print(1)
elif result < 0:
    print(-1)
else:
    print(0)
```

##### [문제] 다각형의 너비 구하기



```python
n = int(input())  # 점의 개수
x = []            # x좌표
y = []            # y좌표


# 좌표 입력
for _ in range(n):
    tmpx, tmpy = map(int, input().split())
    x.append(tmpx)
    y.append(tmpy)
   
# 첫번째 점을 마지막에 추가 
x.append(x[0])
y.append(y[0])

# 넓이 구하기
x_i = 0
y_i = 0
result = 0

# ccw 공식
for i in range(n):
    x_i += x[i] * y[i+1]   # 
    y_i += x[i+1] * y[i]   # 
    result = x_i - y_i
    
print(round(abs(result)/2, 1))  # 절대값으로 바꾸고 2로 나누기
```

#### 슬라이딩 윈도우
- 고정 사이즈의 윈도우가 이동하면서 윈도우 내에 있는 데이터를 이용해 문제를 풀이하는 알고리즘
- 교집합의 정보를 공유하고, 차이가 나는 양쪽 끝 원소만 갱신하는 방법



#### DNA 비밀번호(P65)


```python
# 함수
# 1. 문자 더하기
def addChar(c):  # 새로운 문자(c) 처리 함수
    global chkList, userList, chkSecret
    if c == 'A':
        userList[0] += 1
        if userList[0] == chkList[0]:
            chkSecret += 1
    elif c == 'C':
        userList[1] += 1
        if userList[1] == chkList[1]:
            chkSecret += 1
    elif c == 'G':
        userList[2] += 1
        if userList[2] == chkList[2]:
            chkSecret += 1
    elif c == 'T':
        userList[3] += 1
        if userList[3] == chkList[3]:
            chkSecret += 1
            
# 2. 문자 빼기
def subChar(c):  # 이전 문자(c) 처리 함수
    global chkList, userList, chkSecret
    if c == 'A':
        if userList[0] == chkList[0]:
            chkSecret -= 1
        userList[0] -= 1
    elif c == 'C':
        if userList[1] == chkList[1]:
            chkSecret -= 1
        userList[1] -= 1
    elif c == 'G':
        if userList[2] == chkList[2]:
            chkSecret -= 1
        userList[2] -= 1
    elif c == 'T':
        if userList[3] == chkList[3]:
            chkSecret -= 1
        userList[3] -= 1

```

```python
# 1. chkList 비밀번호 체크 리스트
chkList =[0, 0, 0, 0]
# 2. userList 현재 상태를 저장할 리스트
userList = [0] * 4
# 3. chkSecret 문자의와 관련된 개수 충족 판단 변수
chkSecret = 0
print(chkList, userList, chkSecret)
```

```python
# 구현 코드
# S:문자열클기, P(부분 문장열 크기)
S, P = map(int, input().split())
# A:사용 생성 문자열
A = list(input())

result = 0 # 결과값
# chkList 데이터 입력 받기
chkList=list(map(int, input().split()))
# chkList 비밀번호 만족시(0) chkSecret 값 증가

print(chkList, userList, chkSecret)
```

```python
# p 범위(0~p-1) 만큼 userList/chkSecret 적용, 비밀번호 유효성 판단
# for 사용(p~s)해 반복
#  함수를 이용해 한칸씩 이동하면서 새로운 문자열 생성 및 처리
#  유효한 비밀번호인지 판단 = 결과값 업데이트
for i in range(4):
    if chkList[i] == 0:   # 체크할 리스트중에 0 이 있으면
        chkSecret += 1
        
print(chkList, userList, chkSecret, result)
```

```python
for i in range(P):
    addChar(A[i])
    
print(chkList, userList, chkSecret, result)
```

```python
if chkSecret == 4: # 비밀번호 충족 요건 만족시 만족시
    result += 1 # 결과값 업데이트
    
print(chkList, userList, chkSecret, result)
```

```python
# 위도우 크기 P, 문자열 S => p부터 s까지의 문자열을 슬라이딩 윈도우로 본다.
for i in range(P, S):  # 8 - 9
    j = i - P    # j = 0
    addChar(A[i])
    print(chkList, userList, chkSecret, result)
    
    subChar(A[j])
    print(chkList, userList, chkSecret, result)
    
    if chkSecret == 4:
        result += 1
    
# 결과 출력
print(chkList, userList, chkSecret, result)
```

###### 전체 실행 코드

```python
# 전체 코드

def addChar(c):  # 새로운 문자(c) 처리 함수
    global chkList, userList, chkSecret
    if c == 'A':
        userList[0] += 1
        if userList[0] == chkList[0]:
            chkSecret += 1
    elif c == 'C':
        userList[1] += 1
        if userList[1] == chkList[1]:
            chkSecret += 1
    elif c == 'G':
        userList[2] += 1
        if userList[2] == chkList[2]:
            chkSecret += 1
    elif c == 'T':
        userList[3] += 1
        if userList[3] == chkList[3]:
            chkSecret += 1
            
# 2. 문자 빼기
def subChar(c):  # 이전 문자(c) 처리 함수
    global chkList, userList, chkSecret
    if c == 'A':
        if userList[0] == chkList[0]:
            chkSecret -= 1
        userList[0] -= 1
    elif c == 'C':
        if userList[1] == chkList[1]:
            chkSecret -= 1
        userList[1] -= 1
    elif c == 'G':
        if userList[2] == chkList[2]:
            chkSecret -= 1
        userList[2] -= 1
    elif c == 'T':
        if userList[3] == chkList[3]:
            chkSecret -= 1
        userList[3] -= 1
      
import time  

chkList =[0, 0, 0, 0]
userList = [0] * 4
chkSecret = 0

S, P = map(int, input().split())
A = list(input())
result = 0 
chkList=list(map(int, input().split()))

start_time=time.time()

for i in range(4):
    if chkList[i] == 0:   # 체크할 리스트중에 0 이 있으면
        chkSecret += 1
        
for i in range(P):
    addChar(A[i])
    
if chkSecret == 4: # 비밀번호 충족 요건 만족시 만족시
    result += 1 # 결과값 업데이트
    
for i in range(P, S):  # 8 - 9
    j = i - P    # j = 0
    
    addChar(A[i])
    subChar(A[j])
    
    if chkSecret == 4:
        result += 1
end_time=time.time()
  
# 결과 출력
print(result, end_time-start_time)
```

```python
# 
# 1.  
def addChar(c):  #  (c)  
    global chkList, userList, chkSecret
    if c == 'A':
        userList[0] += 1
        if userList[0] == chkList[0]:
            chkSecret += 1
    elif c == 'C':
        userList[1] += 1
        if userList[1] == chkList[1]:
            chkSecret += 1
    elif c == 'G':
        userList[2] += 1
        if userList[2] == chkList[2]:
            chkSecret += 1
    elif c == 'T':
        userList[3] += 1
        if userList[3] == chkList[3]:
            chkSecret += 1
            
# 2.  
def subChar(c):  #  (c)  
    global chkList, userList, chkSecret
    if c == 'A':
        if userList[0] == chkList[0]:
            chkSecret -= 1
        userList[0] -= 1
    elif c == 'C':
        if userList[1] == chkList[1]:
            chkSecret -= 1
        userList[1] -= 1
    elif c == 'G':
        if userList[2] == chkList[2]:
            chkSecret -= 1
        userList[2] -= 1
    elif c == 'T':
        if userList[3] == chkList[3]:
            chkSecret -= 1
        userList[3] -= 1

```

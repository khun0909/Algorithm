삽입 정렬(insertion sort) 
- 이미 정렬된 데이터 범위에 정렬되지 않은 데이터를 적절한 위치에 삽입시켜 정렬하는 방식
- 시간 복잡도는 O(n^2)으로 느린 편이지만 구현하기가 쉽다



#### ATM 인출시간 계산하기



##### 풀이방법
1. 삽입 정렬을 이용해 인출 시간 Pi를 기준으로 데이터를 오름차순 정렬한다.




2. 정렬된 데이터를 바탕으로 모든 사람이 돈을 인출하는 데 필요한 최솟값을 구한다.
3. 인출에 필요한 시간은 앞사람들의 인출 시간의 합 + 자신의 인출 시간이므로 합 배열로 풀어준다.



```python
# 코드 구현
n = int(input()) # 사람수
A = list(map(int, input().split())) # 각 사람의 인출 시간
S = [0] * n # 각 사람이 돈을 인출하는데 걸리는 시간(누적합)

# 삽입정렬
for i in range(1, n):
    insert_idx = i  # 삽입할 위치
    insert_val = A[i]  # 삽입할 값
    
    # 삽입할 위치를 찾는다.
    for j in range(i-1, -1, -1):
        if A[j] < A[i]:
            insert_idx = j+1
            break
        if j == 0:
            insert_idx = 0
    
    # 삽입
    for j in range(i, insert_idx, -1):
        A[j] = A[j-1]
        
    A[insert_idx] = insert_val
    
S[0]=A[0] # 첫번째 사람은 인출시간이 그대로 걸림

# 누적합
for i in range(1, n): 
    S[i] = S[i-1] + A[i]  
    
sumT =0 # 총 걸린 시간

for i in range(n): # 각 사람이 돈을 인출하는데 걸리는 시간(누적합)의 합
    sumT += S[i]

print(sumT)
```

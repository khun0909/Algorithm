```python
#  
n = int(input()) # 
A = list(map(int, input().split())) #    
S = [0] * n #      ()

# 
for i in range(1, n):
    insert_idx = i  #  
    insert_val = A[i]  #  
    
    #   .
    for j in range(i-1, -1, -1):
        if A[j] < A[i]:
            insert_idx = j+1
            break
        if j == 0:
            insert_idx = 0
    
    # 
    for j in range(i, insert_idx, -1):
        A[j] = A[j-1]
        
    A[insert_idx] = insert_val
    
S[0]=A[0] #     

# 
for i in range(1, n): 
    S[i] = S[i-1] + A[i]  
    
sumT =0 #   

for i in range(n): #      () 
    sumT += S[i]

print(sumT)
```

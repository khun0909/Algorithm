#### 코드 구현

```python
import time
n = 1260
count = 0

# 큰 단위의 화폐부터 차례대로 확인하기
array = [500, 100, 50, 10]

startT = time.time()

for coin in array:
    count += n // coin # 해당 화폐로 거슬러 줄 수 있는 동전의 개수 세기
    n %= coin # 화페로 나눈 나머지 n에 대입
    
endT= time.time()
print(endT - startT)
print(count)
```

#### 슬라이딩 윈도우 알고리즘 문제

```python
n, x = map(int, input().split())
a = list(map(int, input().split()))

curV = 0
maxV = 0
maxC = 0

for i in range(n-x+1):
    if i == 0:
        for j in range(x):
            curV += a[j]
        maxV = curV
        maxC = 1    
    else:
        curV = curV - a[i-1] + a[i+x-1]
        if curV == maxV:
            maxC += 1
        elif curV > maxV:
            maxV = curV
            maxC = 1

print(maxV,'\n', maxC)
```

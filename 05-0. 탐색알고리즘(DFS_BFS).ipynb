    - ! = 1  2  3    ( - 1)  
    -  0! 1!  1


```python
#   n!
def factorial_iterative(n):        
    result = 1
    # 1 n   
    for i in range(1, n + 1):
       result *= i
    return result

#   n!
def factorial_recursive(n):        
    if n <= 1: # n 1   1 
        return 1
    # n! = n * (n - 1)!   
    return n * factorial_recursive(n - 1)

#    n! (n = 5)
print(' :', factorial_iterative(5))
print(' :', factorial_recursive(5))
```

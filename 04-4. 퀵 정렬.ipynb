##### 퀵 정렬(quick sort) 
- 기준값(pivot)을 선정해 해당 값보다 작은 데이터와 큰 데이터로 분류하는 것을 반복해 정렬하는 알고리즘
- 기준값이 어떻게 선정되는지가 시간 복잡도에 많은 영향을 미치고, 평균적인 시간 복잡도는 O(nlogn)이며<br> 최악의 경우에는 시간 복잡도가 O(n^2)이 된다.

##### K번째 수 구하기



- 풀이 과정



```python
#코딩 작업
N, K = map(int, input().split())
data = list(map(int, input().split()))
         
# 위치 바꾸기   
def swap(i, j):
    global data
    data[i], data[j] = data[j], data[i]


# pivot을 기준으로 작은 값은 왼쪽, 큰 값은 오른쪽으로 보내기
def partition(S, E):
    # S: 시작 인덱스, E: 끝 인덱스
    global data
    
    if S+1 == E:
        if data[S] > data[E]: # S가 E보다 크다면
            swap(S, E)       
        return E # pivot의 인덱스 반환
    
    M = (S + E) // 2  # 중간 인덱스
    # print('기준값위치', M)
    swap(S, M)  # S와 M 위치 바꾸기
    pivot = data[S]  # pivot은 S번째 값으로 설정
    # print('기준값', pivot)
    i = S + 1   # 왼쪽에서 오른쪽으로 이동하는 인덱스
    j = E       # 오른쪽에서 왼쪽으로 이동하는 인덱스
    
    while i <= j:
        while pivot < data[j] and j > 0: # pivot보다 큰 값이 나올 때까지
            j -= 1
        
        while pivot > data[i] and i <len(data)-1: # pivot보다 작은 값이 나올 때까지
            i += 1
            
        if i <= j:  
            swap(i, j)
            i += 1
            j -= 1
    # i == j 기준값을 양쪽으로 분리한 가운데에 오도록 설정
    data[S] = data[j]
    data[j] = pivot
    return j

# 퀵 정렬
def quick_sort(S, E, K):
    global data
    
    if S < E:  
        pivot = partition(S, E) # pivot을 기준으로 분할하는 함수 호출
        
        if K == pivot:  # pivot이 K번째라면
            return data[K]
        
        elif K < pivot:  # pivot보다 K가 작다면 왼쪽 그룹에서 찾기
            quick_sort(S, pivot-1, K)  # 재귀호출
            
        else:  # pivot보다 K가 크다면 오른쪽 그룹에서 찾기
            quick_sort(pivot+1, E, K)  

quick_sort(0, N-1, K-1)  # 5 2 => (0, 4, 1)
print(data[K-1])
```

```python
#  

n = int(input()) #  
visited = [False] * (n+1) #    
tree = [[] for _ in range(n+1)] #  
answer = [0] * (n+1) #  

#  
for _ in range(n-1):
    node1, node2 = map(int, input().split())  # 1, 2 
    tree[node1].append(node2) #  
    tree[node2].append(node1) #  
    
# DFS 
def dfs(num):
    visited[num] = True #   
    for i in tree[num]: #   
        if not visited[i]: #  
            answer[i] = num #     
            dfs(i) # 
            
dfs(1) # 1  

for i in range(2, n+1): # 2  
    print(answer[i]) #  
```
